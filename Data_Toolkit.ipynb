Q(2)How does broadcasting work in NumPy?
- NumPy Broadcasting
Broadcasting in NumPy allows arrays of different shapes to be used together in arithmetic operations without explicitly replicating data. It automatically expands smaller arrays to match the shape of larger ones, making computations efficient and memory-saving.

- How Broadcasting Works
Matching Dimensions: NumPy aligns arrays by adding extra dimensions (if needed) to make them compatible.
Expanding Dimensions: The smaller array is virtually expanded (not copied) to match the shape of the larger one.
Element-wise Operations: Operations are performed as if both arrays had the same shape.
Broadcasting Rules
For two arrays to be compatible for broadcasting, their dimensions must satisfy the following conditions:

- If the dimensions of the arrays are not the same, NumPy pads the smaller array with ones on the left.
If corresponding dimensions are equal or one of them is 1, they are compatible.
If dimensions do not match and neither is 1, broadcasting fails.
Examples of Broadcasting
Example 1: Scalar and Array

result = arr + 10  # Scalar is broadcasted to match array shape

print(result)  # Output: [11 12 13]
How it works:
The scalar 10 is treated as an array [10, 10, 10], and element-wise addition is performed.

Q(8)What does the term â€œvectorized operationâ€ mean in NumPy?
- A vectorized operation in NumPy refers to performing operations on entire arrays (or vectors) at once, without the need for explicit loops. These operations are executed at C-level speed, making them significantly faster than standard Python loops.

- Why Use Vectorized Operations?
âœ… Faster Execution â€“ Uses optimized C and Fortran routines under the hood.
âœ… More Readable Code â€“ Eliminates unnecessary loops, making code cleaner.
âœ… Memory Efficient â€“ Uses contiguous memory blocks, improving performance.

- list1 = [1, 2, 3, 4, 5]
- list2 = [10, 20, 30, 40, 50]
- result = []

- start_time = time.time()
for i in range(len(list1)):
    result.append(list1[i] + list2[i])
end_time = time.time()

print(result)  # Output: [11, 22, 33, 44, 55]
print(f"Execution Time: {end_time - start_time:.6f} seconds")
- Examples of Vectorized Operations in NumPy
Operation	Loop-Based Approach	Vectorized Approach
Addition	for loop	arr1 + arr2
Subtraction	for loop	arr1 - arr2
Multiplication	for loop	arr1 * arr2
Division	for loop	arr1 / arr2
Power	for loop	arr1 ** 2
Logarithm	math.log(x) in loop	np.log(arr)
Square Root	math.sqrt(x) in loop	np.sqrt(arr)
- Key Takeaways
- Vectorized operations make NumPy much faster than standard Python loops.
- They leverage low-level optimizations (C and Fortran) for high performance.
Best practice: Always use NumPyâ€™s built-in functions instead of loops for numerical computations.

- ğŸ”¹ Why is NumPyâ€™s Array Slicing Faster Than Pythonâ€™s List Slicing?
NumPyâ€™s array slicing is significantly faster and more efficient than Pythonâ€™s list slicing due to several key reasons, including memory efficiency, optimized operations, and better data handling.

- ğŸ”¹ 1ï¸âƒ£ NumPy Uses Contiguous Memory Allocation

Python lists create copies when sliced, which means extra memory is allocated.
NumPy uses views, meaning it does not create a new copy but rather a new reference to the same memory.

- print(array_data)  # Original NumPy array changes

- ğŸ”¹ 3ï¸âƒ£ NumPy Uses Vectorized Operations
Python lists require looping through elements one by one (slow).
NumPy arrays use C-based vectorized operations, meaning computations are done in parallel (fast).
- âœ… Example: NumPy vs. List Computation Speed

```python
#Q(4)How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
# Generate random data
np.random.seed(42)
data = {
    'A': np.random.rand(10),
    'B': np.random.rand(10),
    'C': np.random.rand(10),
    'D': np.random.rand(10)
}

df = pd.DataFrame(data)
corr_matrix = df.corr()
print(corr_matrix)
# Create a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)

# Add a title
plt.title("Correlation Matrix Heatmap")
plt.show()



```

```python
#Q(8)Create a line plot with multiple lines using Matplotlib?
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 10, 100)  # 100 values between 0 and 10
y1 = np.sin(x)               # Sine wave
y2 = np.cos(x)               # Cosine wave
y3 = np.tan(x)               # Tangent wave (clipped for better visualization)
y3[y3 > 5] = np.nan  # Remove very large values for a clean plot
y3[y3 < -5] = np.nan
plt.figure(figsize=(8, 6))

plt.plot(x, y1, label="Sine Wave", color='blue', linestyle='-')
plt.plot(x, y2, label="Cosine Wave", color='red', linestyle='--')
plt.plot(x, y3, label="Tangent Wave", color='green', linestyle=':')

# Add labels, title, and legend
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Multiple Line Plot")
plt.legend()

# Show the plot
plt.show()


```

```python
#Q(10)Create a histogram using Seaborn to visualize a distribution?
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(42)  # For reproducibility
data = np.random.randn(200)  # 200 random numbers from a normal distribution
plt.figure(figsize=(8, 6))
sns.histplot(data, bins=20, kde=True, color="blue")

# Add labels and title
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of Random Data")

# Show the plot
plt.show()


```

```python
#Q(13)Create a 3D scatter plot using Plotly?
!pip install plotly
import plotly.express as px
import pandas as pd
import numpy as np
np.random.seed(42)  # For reproducibility
df = pd.DataFrame({
    "X": np.random.randn(100),
    "Y": np.random.randn(100),
    "Z": np.random.randn(100),
    "Category": np.random.choice(["A", "B", "C"], size=100)  # Categorical labels
})
fig = px.scatter_3d(df, x="X", y="Y", z="Z",
                    color="Category",  # Color points by category
                    title="3D Scatter Plot",
                    opacity=0.8)

fig.show()



```
